```python
import numpy as np 
import pandas as pd 
#import os
from pyomo.environ import *
print('Libraries were improted successfully!')

T= [1, 2]
B=[1, 2]
P= [1, 2, 3]
model = ConcreteModel()
model.build = Var(B, bounds=(0, None))
model.p=Var(P,T,bounds=(0,None))
model.obj = Objective(expr=100*model.build[1]+150*model.build[2]+200*(9*model.p[1,1]+model.p[1,2])+
                     30*(9*model.p[2,1]+model.p[2,2])+50*(9*model.p[3,1]+model.p[3,2]))

model.constraints=ConstraintList()
model.constraints.add(model.p[1,1]+model.p[2,1]+model.p[3,1]==90)
model.constraints.add(model.p[1,2]+model.p[2,2]+model.p[3,2]==100)
model.constraints.add(model.p[3,1]<=90)
model.constraints.add(model.p[3,2]<=90)
model.constraints.add(model.p[1,1]<=model.build[1])
model.constraints.add(model.p[1,2]<=model.build[1])
model.constraints.add(model.p[2,1]<=model.build[2])
model.constraints.add(model.p[2,2]<=model.build[2])


opt = SolverFactory('cplex', executable=r'/Users/mostafa/Applications/IBM/ILOG/CPLEX_Studio1263/cplex/bin/x86-64_osx/cplex')

# Tell the model that you need the dual values too!
#model.dual = Suffix(direction=Suffix.IMPORT)

# Solve the problem
Solution=opt.solve(model)

# Report the summary of solve
Solution.write(num=1)

print('MW capacity built diesel: ', model.build[1].value)
print('MW capacity built gas: ', model.build[2].value)
print('MW Production for diesel: ', model.p[1,1].value, ' and, ',model.p[1,2].value)
print('MW Production for gas: ', model.p[2,1].value, ' and, ',model.p[2,2].value)













```
