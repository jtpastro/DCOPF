# Distribuição eficiente de energia em SmartGrids 
## Motivação e definição do problema

Nossa sociedade depende cada vez mais do fornecimento energia elétrica devido à evolução tecnológica em diversas áreas. Apesar disso, os sistemas tradicionais de geração e distribuição de eletricidade possuem diversas fragilidades e pouca capacidade de autorrecuperação. Aliado a isso, a adoção de modalidades distribuídas e intermitentes de produção de energia, como fontes eólicas e solares aumentam a complexidade desses sistemas. 
    
Para atender a essas damandas foram propostas técnicas de smart grids. Smart grids são sistemas de controle de geração, transmissão e distribuição de energia capazes de coletar, analisar e adaptar proativamente o estado de uma rede elétrica. Um dos desafios desses sistemas é a recuperação automática do serviço em caso de falhas. Para isso, as smart grids devem ser capazes de redistribuir o fluxo de eletricidade em caso de falhas, minimizando o tempo de indisponibilidade e maximizando a eficiência, tanto energética, quanto econômica.
A proposta desse trabalho é definir e implementar um modelo simples, para distribuição eficiente de energia que atenda das demandas dos usuários, respeite limites físicos do sistema e reduza o custo econômico de distribuição da eletricidade.

Sistemas de geração, transmissão e distribuição de energia são modelados usando diagramas de barramentos de sistema elétrico (power system bus diagram). Todos os elementos dos sistema possuem limitações físicas quanto a voltagem, ângulo de fase, potência e corrente. Esse modelo pode ser simplificado focando na capacidade de produção dos geradores, demanda dos consumidores, restrições elétricas e custo das linhas de transmissão. Ao focar as restrições e custos nas linhas de transmissão, temos que os nós do grafo tornam-se homogêneos, diferenciando-se apenas pela sua relação de oferta/demanda de energia.
